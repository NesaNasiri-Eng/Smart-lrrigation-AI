```python
import random

def get_irrigation_decision(sensor_moisture, temperature, is_raining):
    """
    This function analyzes sensor data and weather to make a decision.
    """
    print(f"--- Processing Data for Location: Graz, Austria ---")
    print(f"Data: Moisture:{sensor_moisture}% | Temp:{temperature}°C | Raining:{is_raining}")

    # Logic for Decision Making
    if is_raining:
        return "SKIP: It is raining. No irrigation needed. 🌧️"

    if sensor_moisture < 30:
        if temperature > 25:
            return "ACTION: Critical! High Temp & Low Moisture. Start Maximum Irrigation. 🔥💧"
        else:
            return "ACTION: Low Moisture. Start Standard Irrigation. ✅"

    if 30 <= sensor_moisture <= 60:
        return "HOLD: Moisture level is optimal. No action required. ⏸️"

    return "WARNING: Soil is too wet. Risk of root rot! ⚠️"

# --- SIMULATION ---
# Instead of manual entry, we simulate the sensor and weather
current_moisture = random.randint(15, 75)
current_temp = 22 # This can be fetched from an API automatically later
raining_now = False

# Get the Result
decision = get_irrigation_decision(current_moisture, current_temp, raining_now)
print(f"Result: {decision}")
```
