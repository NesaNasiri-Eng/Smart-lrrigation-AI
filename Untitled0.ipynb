```python
import requests

# 1. Fetch Real-time Weather Data from Graz, Austria
city = "Graz"
url = f"https://wttr.in/{city}?format=%t+%h+%p"

try:
    response = requests.get(url)
    weather_data = response.text.split()

    # 2. Data Cleaning: Extracting numeric values from strings
    # We remove non-digit characters to get pure integers for Temp and Humidity
    temp = int(''.join(filter(lambda x: x.isdigit() or x == '-', weather_data[0])))
    humidity = int(''.join(filter(lambda x: x.isdigit(), weather_data[1])))
    is_raining = "mm" in weather_data[2]

    # 3. Decision Making Logic (The "Brain" of the System)
    print(f"--- System Status: {city} ---")
    print(f"Current Temp: {temp}C | Humidity: {humidity}% | Raining: {is_raining}")
    print("-" * 35)

    if is_raining:
        print("AI DECISION: [SKIP] - Natural rainfall detected. Conservation mode active.")
    elif temp > 20 and humidity < 50:
        print("AI DECISION: [START] - High evaporation risk. Irrigation activated.")
    else:
        print("AI DECISION: [STANDBY] - Soil moisture levels are optimal.")

except Exception as e:
    print(f"System Error: Unable to fetch live data. Detail: {e}")
```
